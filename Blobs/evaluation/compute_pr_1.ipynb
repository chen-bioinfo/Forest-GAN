```python
import os
import time
import copy
import torch
import random
import numpy as np
from pandas import DataFrame
from method import block_EuclideanDistance, EuclideanDistance, evaluate, evaluation_precision_recall
import matplotlib.pyplot as plt
```

```python
num_disc = 1
lambd = 3
topk = 5

layers_list = [[1,1,1,1], [1,1,2,1], [1,1,2.5,2], [1,1,5,2],[1,1,5,4],[1,1,10,5]]

for layer in layers_list:
    
    recall_list = []
    precision_list = []
    
    train_filename = "train_dataset.npy"
    fake_path = "./fake_sample/small_samples(10000)"
    fake_data_path = os.path.join(fake_path, "GAN_num_disc_1-{}-{}-{}-{}".format(layer[0], layer[1], layer[2], layer[3]))

    print("[precision, recall]")
    for i in range(1, 301):
        start_time = time.time()
        fake_data_filename = os.path.join(fake_data_path, "GAN_epoch={}.npy".format(i*100))
        if not os.path.exists(fake_data_filename):
            break
            
        precision, recall = evaluation_precision_recall(train_filename, fake_data_filename, lambd=lambd, topk=topk)
 
        precision_list.append(precision)
        recall_list.append(recall)
        end_time = time.time()
        print("epoch-{:3d}; precision: {:.5f}; recall: {:.5f}; Times:{:.5f}".format(i, precision, recall, end_time-start_time))
        
    result_pre = "./precision_recall/small_samples(10000)/"

    precision_filename = os.path.join(result_pre, "GAN_num_disc_1-{}-{}-{}-{}_precision".format(layer[0], layer[1], layer[2], layer[3]))
    recall_filename = os.path.join(result_pre, "GAN_num_disc_1-{}-{}-{}-{}_recall".format(layer[0], layer[1], layer[2], layer[3]))

    np.save(precision_filename, np.array(precision_list))
    np.save(recall_filename, np.array(recall_list))
```
